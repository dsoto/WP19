```python
%matplotlib inline
import pandas as pd
import numpy as np
import WP19_analysis as wpa
import seaborn as sns
import matplotlib.pyplot as plt
```

These plots show the cumulative distribution of the blackout durations.
For a given blackout duration on the x-axis, the y-axis tells you the probability that a given blackout event will be that duration or shorter.

The cumulative distribution plots show in the grid-connected regions (Ajau and Asei) only about 5% of the blackout durations are more than three hours.

In the microgrid villages (Atamali, Ayapo, and Kensio), you see the schedule evening outages at around 16 hours and also what appear to be multi-day outages.

```python
data = {}
for rfd in wpa.raw_file_data:
    # get durations from message file
    messages = wpa.load_message_file(rfd['village_name'] + '-messages.csv')
    durations = wpa.get_durations(messages)
    fig, ax = plt.subplots()
    ax.set_title(rfd['village_name'])
    sns.distplot(durations['durations'], 
                 hist_kws=dict(cumulative=True), 
                 kde=False,
                 norm_hist=True,
                 bins=100, 
                 ax=ax)
```
